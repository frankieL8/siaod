### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
user_m = input()
user_n = input()
user_min_limit = input()
user_max_limit = input()
user_m, user_n, user_min_limit, user_max_limit
m = int(user_m)
n = int(user_n)
min_limit = int(user_min_limit)
max_limit = int(user_max_limit)

def matrix_rand(m, n, min_limit, max_limit):
    matrix = []
    for i in range (m):
        array=list()
        for j in range(n):
            array.append(random.randint(min_limit, max_limit))
        matrix.append(array) 
    return matrix

def matrix_print(matrix):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[i])):
            print(matrix[i][j], end=' ')
        print()

matrix = matrix_rand(m, n, min_limit, max_limit)
matrix_print(matrix)
```

Реализовать методы сортировки строк числовой матрицы в соответствии с

```python
import copy
```

```python
# Сортировка выбором. 
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
#TODO
mat = copy.deepcopy(matrix)
def sort_sel(line):
    for i in range(len(line)):
        index = i
        for j in range(i + 1, len(line)):
            if line[index] > line[j]:
                index = j
        temp = line[index]
        line[index] = line[i]
        line[i] = temp
            

for i in range(len(mat)):
    sort_sel(mat[i])

sort_sel_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(sort_sel_time))

matrix_print(mat)
```

```python
# Сортировка вставкой.
start_time = time.time()
#TODO
mat = copy.deepcopy(matrix)
def sort_insertion(line):
    for i in range(len(line)):
        a = line[i]
        pos = i
        
        while pos > 0 and line[pos - 1] > a:
            line[pos] = line[pos - 1]
            pos = pos - 1
        line[pos] = a
    return line

for line in mat:
    line = sort_insertion(line)
matrix_print(mat)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком. 
start_time = time.time()
#TODO
mat = copy.deepcopy(matrix)
def bubble_sort(line):
    for i in range (len(line)):
        k = 0
        for j in range(1, len(line)-i):
            if line[j] < line[j-1]:
                temp = line[j]
                line[j] = line[j-1]
                line[j-1] = temp
                k+=1
        if k == 0:
            break
            
for line in mat:
    line = bubble_sort(line)
matrix_print(mat)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
#TODO
mat = copy.deepcopy(matrix)
def shell_sort(line):
    step = int(len(line) / 2)
    while step > 0:
        for i in range(step, len(line), step):
            key = line[i]
            j = i - step
            while j >= 0 and key < line[j]:
                line[j + step] = line[j]
                j = j - step
            line[j + step] = key
        step = int(step / 2)
        
for i in range(len(mat)):
    shell_sort(mat[i])
    
matrix_print(mat)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
mat = copy.deepcopy(matrix)

start_time = time.time()

def partition(line, start, end):
    piv = line[start]
    left = start + 1
    right = end
    while True:
        while line[right] > piv and right > start:
            right -= 1
        while line[left] <= piv and left < end:
            left += 1
        if left >= right:
            break
        temp = line[left]
        line[left] = line[right]
        line[right] = temp
    line[start] = line[right]
    line[right] = piv
    return right

def __quick_sort(line, start, end):
    if end > start:
        piv_index = partition(line, start, end)
        __quick_sort(line, start, piv_index - 1)
        __quick_sort(line, piv_index + 1, end)
   

def quick_sort(line):
    __quick_sort(line, 0, len(line) - 1)
        
        
for i in range(len(mat)):
    quick_sort(mat[i])
        

quick_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(quick_sort_time))

matrix_print(mat)
    
```

```python
# Турнирная сортировка.
start_time = time.time()
#TODO
import heapq

mat = copy.deepcopy(matrix)

start_time = time.time()

def tournament_sort(line, heapsize):
    heap = []
    if len(line) >= heapsize:
        heap = line[:heapsize]
    else:
        heap = line[:]
    heapq.heapify(heap)
    
    winners = [float('-inf')]
    losers = []
    
    i = heapsize
    while i < len(line) or len(heap) > 0:
        heap_top = heapq.heappop(heap)
        if heap_top >= winners[-1]:
            winners.append(heap_top)
        else:
            losers.append(heap_top)
        
        if i < len(line):
            heapq.heappush(heap, line[i])
        i += 1
    
    del winners[0]
    line.clear()
    line.extend(losers)
    line.extend(winners)
    
    if len(losers) > 0:
        tournament_sort(line, heapsize)
    
for i in range(len(mat)):
    tournament_sort(mat[i], 10)

matrix_print(mat)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
